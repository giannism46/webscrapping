```python
def check_The_City(df):
    Limassol= 0
    Larnaca= 0
    Nicosia= 0
    Famagusta= 0
    Pafos= 0
    citylist=[]
    for i in df['The_city']:
        
        citylist=i.split(', ')
#         print(citylist)
        for LENH in range(len(citylist)):
            if citylist[LENH-1]== 'Λευκωσία':
                Nicosia += 1
            if citylist[LENH-1]== 'Λεμεσός':
                Limassol += 1  
            if citylist[LENH-1]== 'Αμμόχωστος':
                Famagusta += 1   
            if citylist[LENH-1]== 'Λάρνακα':
                Larnaca += 1   
            if citylist[LENH-1]== 'Πάφος':
                Pafos += 1   
        
            
    return   Pafos, Limassol, Nicosia, Larnaca,Famagusta
```
